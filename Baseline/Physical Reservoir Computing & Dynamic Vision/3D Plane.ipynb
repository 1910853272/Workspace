```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d.art3d import Poly3DCollection

# ------------------------------------------------------------
# 1) 飞机网格数据（用户提供的 TR / X，已按列 Xb,Yb,Zb 排列）
# ------------------------------------------------------------
# ------------------------------------------------------------
# 1) 148×3 三角连接矩阵  (Matlab → Python 无需减 1，稍后绘图时再 -1)
# ------------------------------------------------------------
TR = np.array([
    [1, 50, 41], [1, 51, 50], [2, 51, 1],  [3, 51, 2],  [3, 53, 51],
    [3, 43, 53], [3, 4, 43],  [4, 3, 52],  [5, 4, 52],  [5, 47, 4],
    [5, 49, 47], [6, 49, 5],  [6, 54, 49], [7, 54, 6],  [8, 54, 7],
    [9, 54, 8],  [9, 16, 54], [9, 10, 16], [10, 9, 28], [11, 10, 28],
    [11, 14, 10], [11, 13, 14], [12, 13, 11], [13, 12, 24], [13, 24, 14],
    [14, 24, 20], [15, 14, 20], [15, 10, 14], [16, 10, 15], [15, 48, 16],
    [17, 48, 15], [17, 37, 48], [17, 36, 37], [18, 36, 17], [18, 8, 36],
    [19, 8, 18],  [9, 8, 19],  [9, 19, 28], [19, 55, 28], [19, 35, 55],
    [19, 18, 35], [18, 20, 35], [18, 17, 20], [17, 15, 20], [20, 24, 35],
    [21, 35, 24], [22, 35, 21], [22, 34, 35], [22, 28, 34], [22, 29, 28],
    [22, 23, 29], [23, 22, 21], [23, 21, 56], [21, 33, 56], [21, 24, 33],
    [24, 25, 33], [12, 25, 24], [25, 12, 11], [25, 11, 26], [26, 32, 25],
    [26, 27, 32], [27, 26, 30], [26, 29, 30], [11, 29, 26], [28, 29, 11],
    [29, 23, 30], [30, 23, 56], [27, 30, 56], [31, 27, 56], [32, 27, 31],
    [33, 32, 31], [25, 32, 33], [33, 31, 56], [34, 28, 55], [35, 34, 55],
    [36, 8, 7],   [36, 7, 38],  [37, 36, 38], [38, 46, 37], [38, 39, 46],
    [39, 38, 7],  [40, 39, 7],  [39, 40, 1],  [1, 45, 39], [1, 41, 45],
    [41, 42, 45], [42, 41, 43], [43, 4, 42],  [42, 4, 44], [42, 44, 45],
    [44, 47, 45], [4, 47, 44],  [45, 47, 46], [45, 46, 39], [46, 47, 37],
    [37, 47, 48], [47, 49, 48], [48, 49, 16], [49, 54, 16], [41, 50, 43],
    [50, 53, 43], [51, 53, 50], [1, 40, 2],  [2, 40, 3],  [3, 40, 52],
    [52, 40, 7],  [6, 52, 7],  [6, 5, 52],  [57, 60, 59], [57, 58, 60],
    [58, 57, 61], [57, 62, 61], [57, 59, 62], [59, 58, 62], [60, 58, 59],
    [61, 62, 58], [63, 64, 66], [64, 63, 68], [64, 68, 67], [65, 64, 67],
    [66, 64, 65], [63, 66, 65], [63, 65, 67], [67, 68, 63], [69, 71, 73],
    [70, 71, 69], [71, 70, 74], [70, 73, 74], [70, 69, 73], [72, 74, 73],
    [72, 71, 74], [71, 72, 73], [75, 76, 79], [76, 75, 77], [77, 75, 80],
    [75, 79, 80], [78, 80, 79], [78, 76, 80], [76, 78, 79], [76, 77, 80],
    [81, 82, 85], [82, 81, 83], [83, 81, 86], [81, 85, 86], [84, 86, 85],
    [84, 82, 86], [82, 84, 85], [82, 83, 86]
], dtype=int)

# ------------------------------------------------------------
# 2) 86×3 三维顶点坐标 (机体坐标系)
# ------------------------------------------------------------
X_b = np.array([
    [ 0.4 ,  0.025 , -0.0433012701892219],
    [ 0.4 ,  0.05  ,  0.0                ],
    [ 0.4 ,  0.025 ,  0.0433012701892219],
    [ 0.4 , -0.025 ,  0.0433012701892219],
    [ 0.3 , -0.035 ,  0.0606217782649107],
    [ 0.2 ,  0.035 ,  0.0606217782649107],
    [ 0.2 ,  0.07  ,  0.0               ],
    [ 0.0 ,  0.065 ,  0.0               ],
    [-0.1 ,  0.03  ,  0.0519615242270663],
    [-0.1 , -0.03  ,  0.0519615242270663],
    [-0.2 , -0.025 ,  0.0433012701892219],
    [-0.3 , -0.04  ,  0.0               ],
    [-0.2 , -0.05  ,  0.0               ],
    [-0.1 , -0.06  ,  0.0               ],
    [ 0.0 , -0.065 ,  0.0               ],
    [ 0.0 , -0.0325,  0.0562916512459885],
    [ 0.0 , -0.0325, -0.0562916512459885],
    [-0.1 ,  0.03  , -0.0519615242270663],
    [-0.1 ,  0.06  ,  0.0               ],
    [-0.1 , -0.03  , -0.0519615242270663],
    [-0.4 ,  0.015 , -0.0259807621135332],
    [-0.4 ,  0.03  ,  0.0               ],
    [-0.5 ,  0.02  ,  0.0               ],
    [-0.2 , -0.025 , -0.0433012701892219],
    [-0.4 , -0.03  ,  0.0               ],
    [-0.4 , -0.015 ,  0.0259807621135332],
    [-0.5 , -0.01  ,  0.0173205080756888],
    [-0.2 ,  0.025 ,  0.0433012701892219],
    [-0.4 ,  0.015 ,  0.0259807621135332],
    [-0.5 ,  0.01  ,  0.0173205080756888],
    [-0.5 , -0.01  , -0.0173205080756888],
    [-0.5 , -0.02  ,  0.0               ],
    [-0.4 , -0.015 , -0.0259807621135332],
    [-0.3 ,  0.04  ,  0.0               ],
    [-0.2 ,  0.025 , -0.0433012701892219],
    [ 0.0 ,  0.0325, -0.0562916512459885],
    [ 0.2 , -0.035 , -0.0606217782649107],
    [ 0.2 ,  0.035 , -0.0606217782649107],
    [ 0.3 ,  0.035 , -0.0606217782649107],
    [ 0.3 ,  0.07  ,  0.0               ],
    [ 0.5 , -0.01  , -0.0173205080756888],
    [ 0.5 , -0.02  ,  0.0               ],
    [ 0.5 , -0.01  ,  0.0173205080756888],
    [ 0.4 , -0.05  ,  0.0               ],
    [ 0.4 , -0.025 , -0.0433012701892219],
    [ 0.3 , -0.035 , -0.0606217782649107],
    [ 0.3 , -0.07  ,  0.0               ],
    [ 0.2 , -0.07  ,  0.0               ],
    [ 0.2 , -0.035 ,  0.0606217782649107],
    [ 0.5 ,  0.01  , -0.0173205080756888],
    [ 0.5 ,  0.02  ,  0.0               ],
    [ 0.3 ,  0.035 ,  0.0606217782649107],
    [ 0.5 ,  0.01  ,  0.0173205080756888],
    [ 0.0 ,  0.0325,  0.0562916512459885],
    [-0.2 ,  0.05  ,  0.0               ],
    [-0.5 ,  0.01  , -0.0173205080756888],
    [ 0.2 ,  0.0657784834550136,  0.0239414100327968],
    [ 0.0 ,  0.0562916512459885, -0.0325             ],
    [ 0.05,  0.5  ,  0.0               ],
    [ 0.2 ,  0.07 ,  0.0               ],
    [ 0.0 ,  0.061080020351084, 0.0222313093161685 ],
    [-0.05,  0.5  ,  0.0               ],
    [ 0.0 , -0.0562916512459885, -0.0325           ],
    [ 0.2 , -0.0657784834550136, 0.0239414100327968],
    [ 0.05, -0.5  ,  0.0               ],
    [ 0.2 , -0.07 ,  0.0               ],
    [-0.05, -0.5  ,  0.0               ],
    [ 0.0 , -0.061080020351084, 0.0222313093161685 ],
    [-0.45,  0.2  ,  0.0               ],
    [-0.4 ,  0.0259807621135332,  0.015],
    [-0.4 ,  0.0259807621135332, -0.015],
    [-0.5 ,  0.0173205080756888, -0.01 ],
    [-0.5 ,  0.2  ,  0.0               ],
    [-0.5 ,  0.0173205080756888,  0.01 ],
    [-0.4 , -0.0259807621135332,  0.015],
    [-0.4 , -0.0259807621135332, -0.015],
    [-0.45, -0.2  ,  0.0               ],
    [-0.5 , -0.0173205080756888, -0.01 ],
    [-0.5 , -0.0173205080756888,  0.01 ],
    [-0.5 , -0.2  ,  0.0               ],
    [-0.4 ,  0.012678547852221 ,  0.0271892336110995],
    [-0.4 , -0.012678547852221 ,  0.0271892336110995],
    [-0.45,  0.0  ,  0.15 ],
    [-0.5 , -0.00845236523481399, 0.018126155740733 ],
    [-0.5 ,  0.00845236523481399, 0.018126155740733 ],
    [-0.5 ,  0.0  ,  0.15 ]
])

# 为了在大型场景里更直观，给飞机放大到 1000× (可自由调节)
SCALE = 2_000
X_b *= SCALE

# ------------------------------------------------------------
# 2) 生成螺旋轨迹（radius, pitch, 采样点数可自定义）
# ------------------------------------------------------------
n_pts   = 10                      # 画几架飞机
theta   = np.linspace(0, 2*np.pi, n_pts, endpoint=False)
radius  = 10_000                  # 螺旋半径 (m)
pitch   = 1_000                   # 每圈上升高度 (m)

pos_t   = np.stack([
    radius * np.cos(theta),       # x
    radius * np.sin(theta),       # y
    pitch  * theta               # z
], axis=1)

# ------------------------------------------------------------
# 3) 工具：由三维方向向量求机头朝向的姿态角
# ------------------------------------------------------------
def direction_to_euler(dir_vec, roll=0):
    """由 (dx,dy,dz) 得到 (yaw, pitch, roll)（右手系）
    yaw   : 绕 +Z 轴旋转 (heading)，机头右转为负，左转为正
    pitch : 绕 +Y 轴旋转 (抬头为正)
    roll  : 自定义（可固定，也可自行设定规律）
    返回的三个角度按 Rz·Ry·Rx 顺序使用
    """
    dx, dy, dz = dir_vec / np.linalg.norm(dir_vec)
    yaw   = np.arctan2(dy, dx)
    pitch = np.arctan2(-dz, np.sqrt(dx*dx + dy*dy))
    return yaw, pitch, roll

def euler_matrix(yaw, pitch, roll):
    """R = Rz(yaw)·Ry(pitch)·Rx(roll)"""
    cy, sy = np.cos(yaw),   np.sin(yaw)
    cp, sp = np.cos(pitch), np.sin(pitch)
    cr, sr = np.cos(roll),  np.sin(roll)

    Rz = np.array([[cy, -sy, 0],
                   [sy,  cy, 0],
                   [ 0,   0, 1]])
    Ry = np.array([[ cp, 0, sp],
                   [  0, 1,  0],
                   [-sp, 0, cp]])
    Rx = np.array([[1,  0,   0],
                   [0, cr, -sr],
                   [0, sr,  cr]])
    return Rz @ Ry @ Rx

# ------------------------------------------------------------
# 4) 为每个采样点复制飞机、姿态变换、绘制
# ------------------------------------------------------------
fig = plt.figure(figsize=(7, 6))
ax  = fig.add_subplot(111, projection='3d')
ax.set_box_aspect((1,1,1))

all_poly = []

for k in range(n_pts):
    # 计算该段切向（朝前方向向量）
    if k == n_pts - 1:
        dir_vec = pos_t[0] - pos_t[k]
    else:
        dir_vec = pos_t[k+1] - pos_t[k]

    yaw, pitch, roll = direction_to_euler(dir_vec, roll=0.0)   # 保持水平无横滚
    R = euler_matrix(yaw, pitch, roll)

    # 机身坐标 -> 世界坐标：R·X_b.T + pos
    verts_w = (R @ X_b.T).T + pos_t[k]

    # 收集三角面
    faces = [[verts_w[i-1], verts_w[j-1], verts_w[k_-1]]  # TR 从 1 开始
             for i, j, k_ in TR]

    color = "#d62728" if k % 2 == 0 else "#1f77b4"        # 间隔上色
    poly  = Poly3DCollection(faces, facecolor=color, edgecolor='k', linewidths=0.2, alpha=0.9)
    ax.add_collection3d(poly)
    all_poly.append(poly)

# 画轨迹（红-蓝两条虚线示意）
ax.plot(pos_t[:,0], pos_t[:,1], pos_t[:,2], '--', color='r', linewidth=1.5, alpha=0.6)
#ax.plot(pos_t[:,0], pos_t[:,1], pos_t[:,2]+500, '--', color='b', linewidth=1.5, alpha=0.6)

# ------------------------------------------------------------
# 5) 坐标轴&视图调优
# ------------------------------------------------------------
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_xlim(-radius*1.2, radius*1.2)
ax.set_ylim(-radius*1.2, radius*1.2)
ax.set_zlim(0, pos_t[:,2].max()*1.1)
ax.view_init(elev=20, azim=-60)      # 俯仰/方位视角，可再调
plt.tight_layout()
plt.show()
```
